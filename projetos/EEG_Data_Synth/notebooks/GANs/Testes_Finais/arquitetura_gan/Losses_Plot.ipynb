```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Path to the folder containing the PNG plots
folder_path = "./losses_for_plot"

# Get a list of all PNG files in the folder
plot_files = [file for file in os.listdir(folder_path) if file.endswith(".png")]

# Create a 3x3 grid
fig, axs = plt.subplots(3, 3, figsize=(12, 12))

# Loop through the plot files and display them in the grid
for i, plot_file in enumerate(plot_files):
    # Calculate the row and column indices for the subplot
    row_index = i // 3
    col_index = i % 3
    
    # Load the image and display it in the subplot
    img = mpimg.imread(os.path.join(folder_path, plot_file))
    axs[row_index, col_index].imshow(img)
    axs[row_index, col_index].axis('off')  # Turn off axis labels for cleaner appearance

# Adjust layout to prevent clipping of titles
plt.tight_layout()

plt.savefig("grid_plot_loss.png")
# Show the plot
plt.show()
```
